# ***Remove URLs***

```python
import re
```

```python
def remove_url(text):
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub("",text)
```

```python
print(data.columns)
```

```python
data['review'] = data['review'].apply(remove_url)
```

# ***Remove Punctuations***

```python
import string

string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,"")
    return text
```

```python
data['review'] = data['review'].apply(remove_punc)
```

```python
chat_words={
    "LOL": "Laugh Out Loud",
    "BRB": "Be Right Back",
    "TTYL": "Talk To You Later",
    "BTW": "By The Way",
    "FYI": "For Your Information",
    "OMG": "Oh My God",
    "IDK": "I Don‚Äôt Know",
    "IMO": "In My Opinion",
    "IMHO": "In My Humble Opinion",
    "IRL": "In Real Life",
    "BFF": "Best Friends Forever",
    "SMH": "Shaking My Head",
    "TBH": "To Be Honest",
    "ROFL": "Rolling On the Floor Laughing",
    "TMI": "Too Much Information",
    "NP": "No Problem",
    "DM": "Direct Message",
    "ASAP": "As Soon As Possible",
    "GG": "Good Game",
    "TL;DR": "Too Long; Didn‚Äôt Read",
    "IKR": "I Know, Right?",
    "FWIW": "For What It‚Äôs Worth",
    "AFK": "Away From Keyboard",
    "NVM": "Never Mind",
    "YOLO": "You Only Live Once"
}
```

```python
chat_words["ASAP"]
```

```python
def chat_conversion(text):
    new_text=[]
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
data['review'] = data['review'].apply(chat_conversion)
```

# ***Stop word removal***

```python

import nltk
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords.words("english")
```

```python
def remove_stopwords(text):
    new_text=[]

    for word in text.split():
        if word in stopwords.words("english"):
            new_text.append("")
        else:
            new_text.append(word.strip())


    return " ".join(new_text).replace("   ","")
```

```python
import spacy

# Load the English NLP model from spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def tokenize_text(text):
    # Process the text with the spacy model
    doc = nlp(text)

    # Extract tokens from the processed document
    tokens = [token.text for token in doc]

    return tokens
```

# ***Stemming and Lemmatization***

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj=PorterStemmer()

    stem_word=[obj.stem(word) for word in text.split()]

    return stem_word
```

```python
stemming(data['review'][217042])
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

def lammatization(text):
    words=text.split()

    lemmetizer=WordNetLemmatizer()

    lemetized_word=[lemmetizer.lemmatize(word) for word in words]

    return lemetized_word
```
